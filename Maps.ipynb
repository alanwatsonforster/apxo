## Duplicate Map Sheets

We avoid copies of A2 (Hanoi, Paul Doumer bridge, and Red River) and B1 (Canale bridges and railyard) as they contain prominent features.

- Based on A1:
  - A3 = A1 rotated simplified
  - A4 = A1 rotated simplified
  - A5 = A1 upright simplified
  - A6 = A1 rotated simplified
- Based on B2:
  - B3 = B2 rotated
  - B4 = B2 rotated (for V-11)
  - B3 = B2 upright
  - B4 = B2 rotated
- Based on C1 and C2:
  - C3 = C1 rotated (for V-22)
  - C4 = C2 rotated (for V-22)
  - C5 = C1 upright
  - C6 = C2 rotated simplified

Simplified means runways, taxiways, dams, lakes, towns with more than 5 hexes, and cities are removed. This has no effect on B2 or C1 as they do not contain these features.

## Hanoi

This is my best attempt at a map of Hanoi and its surrounding regions, adapted from scenarios V-13, V-23, V-24, and V-25.

Hanoi itself is on A2. In the various scenarios, there are ground units or targets set up in B1 to the north of A2, B2 to the east, A1 to the south, and C2 to the west. Other maps in the scenarios are only occupied by aircraft and so can be freely substituted. To fit all of the scenarios on a rectangular map, we need it to extend two maps north and south from A2 and one map east and west from A2. This is shown in the first image. 

We want to use A3, A4, and A to represent the wide Red River flowing from north west to south east. This is shown in the second image.

We then fill in the missing map sheets from the remaining B and C sheets. We do not use C4, as it has a prominent dam. The final map sheet is shown in the third image. It is 3 sheets (60 hexes or 17 miles) east-to-west and 5 sheets (75 hexes or 25 miles) north-to-south.
